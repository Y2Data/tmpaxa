```python
import pandas as pd
```

```python
df = pd.read_csv('./example_data_2019_v1.csv')
```

```python
def clean_df(df):
    df = df.drop(columns=['Category'])
    df.Action_time = pd.DataFrame.astype(df.Action_time, "datetime64[ns]")
    return df
```

```python
df = clean_df(df)
df
```

```python
allUsers = df.User_ID.unique().tolist()
allDataFrames = [df[df.User_ID == user] for user in allUsers]
tmp = allDataFrames[0]
```

```python
tmp['more_than_5m59s'] = tmp['Action_time'].diff() > pd.Timedelta(seconds=359)
tmp[tmp.more_than_5m59s == True]
```
